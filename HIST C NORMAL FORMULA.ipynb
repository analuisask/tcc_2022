```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error
```

```python
tabelas = ['variaveis']
variaveis = pd.read_excel('Dados_-_novo.xlsx', engine="openpyxl")
print(variaveis)
```

```python
variaveis["NRO_EMP"] = variaveis["NRO_EMP"].replace('0', 1.697)
print(variaveis.isnull().sum())
```

```python
sns.histplot(variaveis['IDHM'], bins=50)
```

```python
sns.histplot(variaveis['POP_EST'], bins=50)
```

```python
sns.histplot(variaveis['MASSA_PCAP'], bins=50)
```

```python
norm_form = variaveis.copy()
```

```python
for col in ['POP_EST','DENS_DEM','IDHM','PIBCAP','MASSA_PCAP','CUSTO_UNIM','MASSA_PCAP_POP','DESP_TOT_RSU']:
    norm_form[col]= (norm_form[col] - norm_form[col].mean()) / norm_form[col].std()
```

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
```

```python
p=sns.histplot(norm_form['POP_EST'], bins=50)
p.set_xlabel("População estimada", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
p.ticklabel_format(style='plain', axis='x')
```

```python
p=sns.histplot(variaveis['DENS_DEM'], bins=50)
p.set_xlabel("Densidade demográfica", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histDENS.pdf')
```

```python
p=sns.histplot(variaveis['IDHM'], bins=50)
p.set_xlabel("IDH do município", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histIDHM.pdf')
```

```python
p=sns.histplot(variaveis['PIBCAP'], bins=40)
p.set_xlabel("PIB per capita no município", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histPIB.pdf')
```

```python
p=sns.histplot(variaveis['MASSA_PCAP'], bins=50)
p.set_xlabel("Massa per capita", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histMASSAPCAP.pdf')
```

```python
p=sns.histplot(variaveis['CUSTO_UNIM'], bins=50)
p.set_xlabel("Custo unitário médio", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histCUSTOUNI.pdf')
```

```python
p=sns.histplot(variaveis['MASSA_PCAP_POP'], bins=50)
p.set_xlabel("Massa per capita em relação à população atendida", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
plt.savefig('histMASSAPCAPOP.pdf')
```

```python
p=sns.histplot(variaveis['DESP_TOT_RSU'], bins=50)
p.set_xlabel("Despesa total com RSU", fontsize = 12)
p.set_ylabel("Frequência", fontsize = 12)
p.ticklabel_format(style='plain', axis='x')
```
